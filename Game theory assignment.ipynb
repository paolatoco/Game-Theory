```python
import random 

# Board game with the points each player wins.
points = [[6, 4], [10, 1]]
    
# Dictionary of points B: Buy the house, N: Not buy the house
dictionary_results = {"(B, B)": (points[0][0], points[0][0]),  #points (6,6)
                          "(B, N)": (points[0][1], points[1][0]),  #points (4,10)
                          "(N, B)": (points[1][0], points[0][1]),  #points (10,4)
                          "(N, N)": (points[1][1], points[1][1])}  #points (1,1)

Score = [0,0]
user_play = input(f"""\nHi there! I'm Paola a famous realtor in Boston and Massachusetts! 

  \_/          
-=(_)=-  /\||   /\||   /\||
  / \   //\\|   //\\|   //\\|
       //  \\  //  \\  //  \\
      //    \^/    \^/    \\
      |[]  []|[]  []|[]  []|
     &|  ||  %  ||  |  ||  |%

These days have been really tough for me... my clients are pretty confused and don't know
if buying or not a house 

Wanna help them decide whether they buy the house or not? (Only valid answers yes - no) """).lower()

if "no" in user_play:
    print(f"\n I'm sorry to hear that you don't want to help my clients decide :(")
    
    print("""
    The situation was the following:

Katy and Sean got married recently and contacted my real
state company to buy a house.
   
I gave them a beautiful option that fullfils all their requirements:
       
       A big house near Rockport, MA with 4 beds and 3 baths, 2 car spots
       and is just 5 min walk to the beach.
 
The scenarios of the game are the following:

                              Sean
                     -----------------------     
                     |Buy House | No Buy   |    
           |---------|--------- |----------|
           |Buy House| both earn| Sean wins|
           |         | 6 points | 10 points|
           |         |          | Katy 4   |
    Katy   |---------|--------- |----------|
           | No Buy  | Katy wins| both earn|
           |         | 10 points| 1 point  |
           |         |  Sean 4  |          |
           ---------------------------------
and the game started!
    """)
    
    # beginngin of the loop for the 'no'
    for n in range(5):

        #the strategy used for the 'no' was to always the second python decided to buy the house
        
        computer_decision = random.randint(0,1) # 0: buy the house 1: not buy it
        computer_decision2 = "B" 

        # conditional for the loop. It gives 5 results.

        if computer_decision2 == "B" and computer_decision == 0:
            user_result = dictionary_results["(B, B)"][0]
            computer_result = dictionary_results["(B, B)"][1]
            Score[0] += user_result
            Score[1] += computer_result

            print(f"""
        You've decided to buy the beautiful house, as well as Sean!
        Katy's points {user_result}.
        Sean earned {computer_result}""")

        elif computer_decision2 == "N" and computer_decision == 0:
            user_result = dictionary_results["(N, B)"][0]
            computer_result = dictionary_results["(N, B)"][1]
            Score[0] += user_result
            Score[1] += computer_result
            print(f"""
        You don't want to buy the house, but Sean wants it!
        Katy's points {user_result}.
        Sean earned {computer_result}""")

        elif computer_decision2 == "B" and computer_decision == 1:
            user_result = dictionary_results["(B, N)"][0]
            computer_result = dictionary_results["(B, N)"][1]
            Score[0] += user_result
            Score[1] += computer_result
            print(f"""
        You've decide to buy the house, but not Sean.
        Katy's points {user_result}.
        Sean earned {computer_result}""")

        elif computer_decision2 == "N" and computer_decision == 1:
            user_result = dictionary_results["(N, N)"][0]
            computer_result = dictionary_results["(N, N)"][1]
            Score[0] += user_result
            Score[1] += computer_result
            print(f"""
        We both decided to not buy the house
        Katy's points {user_result}.
        Sean earned {computer_result}""")


        # Declaring a winner for the 'No' part

    user_result = Score[0]
    computer_result = Score[1]

    if user_result > computer_result:
        print(f"""Katy won!""")
        print(f"Katy result is: {Score[0]} - Sean result is: {Score[1]}")


    elif user_result == computer_result:
        print(f"""There was a tie!""")
        print(f"Katy result is: {Score[0]} - Sean result is: {Score[1]}")

    elif user_result <= computer_result:
        print(f"""Sean won! Sorry Katy :(""")
        print(f"Katy result is: {Score[0]} - Sean result is: {Score[1]}")


# If the user decides to play the game!
if "yes" in user_play:
    user_name = input("Yey!! You decided to play :) First, whats your name? ").capitalize()
    
    print(f"""\n Hi {user_name} I'm happy that you accepted this challenge :)...
Trust me, you will have fun!!
    """)
    print(f"""
Okay, this is the situation, you and Sean got married recently and contacted my real
state company to buy a house.
   
I gave you a beautiful option that fullfils all the requirements:
       
       A big house near Rockport, MA with 4 beds and 3 baths, 2 car spots
       and is just 5 min walk to the beach.
 
 The scenarios of the game are the following:
   
                                        Sean
                        -------------------------------------     
                        |Buy House        | No Buy          |    
              |---------|-----------------|-----------------|
              |Buy House| both earn       | Sean wins       |
              |         | 6 points        | 10 points       |
              |         |                 | {user_name}   4       |
  {user_name}       |---------|-----------------|-----------------|
              | No Buy  | {user_name} wins      | both earn       |
              |         | 10 points       | 1 point         |
              |         |  Sean 4         |                 |
              -----------------------------------------------

Let's start!!

    """)
    
    # the user decides what to do
    for n in range(5):
        user_choose = input("""
        What would you choose?
        Select:
        - 'B' Buy the house
        - 'N' Not buy the house
        """).upper()
    
        # Computer decides randomly either to buy the house or not
        computer_decision = random.randint(0,1) # 0: buy the house 1: not buy it

        # conditional structure with the outputs of each scenario
        # the while is to not have any errors when answering the previuos input
        while True:
            if user_choose == "B" or user_choose == "N":
                break
                print("The correct answers are B or N")
                user_choose = input().upper()
        if user_choose == "B" and computer_decision == 0:
            user_result = dictionary_results["(B, B)"][0]
            computer_result = dictionary_results["(B, B)"][1]
            Score[0] += user_result
            Score[1] += computer_result

            print(f"""
        You've decided to buy the beautiful house, as well as Sean!
        Your points {user_result}.
        Sean earned {computer_result}""")

        elif user_choose == "N" and computer_decision == 0:
            user_result = dictionary_results["(N, B)"][0]
            computer_result = dictionary_results["(N, B)"][1]
            Score[0] += user_result
            Score[1] += computer_result
            print(f"""
        You don't want to buy the house, but Sean wants it! 
        Your points {user_result}.
        Sean earned {computer_result}""")

        elif user_choose == "B" and computer_decision == 1:
            user_result = dictionary_results["(B, N)"][0]
            computer_result = dictionary_results["(B, N)"][1]
            Score[0] += user_result
            Score[1] += computer_result
            print(f"""
        You've decide to buy the house, but not Sean.
        Your points {user_result}
        Sean earned {computer_result}""")

        elif user_choose == "N" and computer_choose == 1:
            user_result = dictionary_results["(N, N)"][0]
            computer_result = dictionary_results["(N, N)"][1]
            Score[0] += user_result
            Score[1] += computer_result
            print(f"""
        We both decided to not buy the house
        Your points {user_result}.
        Sean earned {computer_result}""")


        # Declaring a winner.

    user_result = Score[0]
    computer_result = Score[1]

    if user_result > computer_result:
        print(f"""Congrats, you won!""")
        print(f"{user_name} Your result is: {Score[0]} - Sean's result is: {Score[1]}")


    elif user_result == computer_result:
        print(f"""You got into a tie with Sean!""")
        print(f"{user_name} Your result is: {Score[0]} - Sean's result is: {Score[1]}")

    elif user_result <= computer_result:
        print(f"""Sean won :(""")
        print(f"{user_name} Your result is: {Score[0]} - Sean's result is: {Score[1]}")
```
